# Problem Understanding

The challenge is to recommend compatible participants in networking events.

Each participant has structured attributes (age, role, seniority, industry, etc.) and unstructured text attributes (interests, objectives, constraints).

Given historical compatibility scores between pairs, the task is to learn a function:

f(user_A, user_B) → compatibility

This is fundamentally a **pairwise similarity / matching problem**, not classical regression.

Hence, the solution is framed as:

> Learning embeddings for users such that compatible users lie close in embedding space.

This naturally motivates a **Siamese / Metric Learning architecture**.

---

# Key Observations

• Compatibility is symmetric  
• Dataset provides only positive examples  
• Test requires full cartesian prediction  
• Features are heterogeneous (numeric + categorical + text)

Therefore:

1. Pairs are mirrored for symmetry
2. Negative pairs are synthetically generated
3. Profiles are encoded into dense vectors
4. Compatibility is learned from embedding interactions

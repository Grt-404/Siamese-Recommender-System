# Feature Engineering

Each user profile consists of:

• Numeric attributes (Age, Company Size)
• Categorical attributes (Role, Industry, City, etc.)
• Free-text attributes (Interests, Objectives, Constraints)

Processing steps:

1. Numeric features standardized (StandardScaler)
2. Categorical features one-hot encoded
3. Text fields merged and vectorized using TF-IDF

Final representation:

Each participant is converted into a single dense feature vector by concatenating:

[numerical | categorical | text]

This allows the neural network to learn directly from raw profile attributes without manual feature crafting.
